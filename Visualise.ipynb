```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
mega=pd.read_csv('marveldc.csv',encoding='unicode_escape')
mega = mega.drop('Unnamed: 0', 1)
dc=mega[23:40]
marvel=mega[0:23]
marvel.at[marvel[marvel['Original Title']=='Captain Marve'].index[0],'Original Title']='Captain Marvel'
marvel.at[marvel[marvel['Original Title']=='Iron Man Three'].index[0],'Original Title']='Iron Man 3'
data_top=mega.head()
```

```python
#Opening Weekend Comparison Marvel v DC
x1=mega["Opening Weekend USA"].groupby(mega["Company"]).sum()
x1 = pd.DataFrame(x1)
heads = ['Company', 'Earnings']
data = [['DC', x1['Opening Weekend USA'][0]], ['Marvel', x1['Opening Weekend USA'][1]]]
x = pd.DataFrame(data, columns=heads)
ax = sns.barplot(x='Company', y='Earnings', data=x, palette="Blues_d").set(title='Opening Weekend')
```

```python
#Gross Worldwide Comparison Marvel v DC
x1=mega["Gross Worldwide"].groupby(mega["Company"]).sum()
x1 = pd.DataFrame(x1)
heads = ['Company','Earnings']
data = [['DC',x1['Gross Worldwide'][0]],['Marvel',x1['Gross Worldwide'][1]]]
x = pd.DataFrame(data, columns=heads)
ax = sns.barplot(x='Company', y='Earnings', data=x, palette="Blues_d").set(title='Gross Worldwide Earnings')
```

```python
#Earning comparison of all Marvel movies
plt.figure()

ax = sns.barplot(x='Gross Worldwide', y='Original Title',data=marvel, palette='CMRmap_r').set(title='Gross Worldwide Earnings')
# marvel["Gross Worldwide"].plot(kind='barh')

# plt.title("Gross Worldwide Earnings")
# plt.xlabel("Earnings in 10^9")
# plt.ylabel("Movie Name")
# plt.legend("$")
```

```python
#Earning comparison of all DC movies
plt.figure()
ax = sns.barplot(x='Gross Worldwide', y='Original Title',data=dc, palette='GnBu').set(title='Gross Worldwide Earnings')
```

```python
#Rating Comparison of all Marvel movies
plt.figure()
ax = sns.barplot(x='Rate', y='Original Title',data=marvel, palette='PuBuGn').set(title='Rating Comparison (Marvel)')
```

```python
#Rating Comparison of all DC movies
plt.figure()
ax = sns.barplot(x='Rate', y='Original Title',data=dc, palette='PuBuGn').set(title='Rating Comparison (DC)')
```

```python
#Release date comparison of all Marvel movies against Worldwide Gross
"""
We can see a general trend in increase and then slight decrease 
followed by increase in popularity of Marvel movies over time which helps 
us to capture people's general sentiment and likings towards the movies
"""
plt.figure()
ax=sns.scatterplot(x="Release",y="Gross Worldwide", data=marvel, hue="Gross Worldwide", size="Gross Worldwide", palette="copper").set(title="Release Date Comparison of All DC Movies")
```

```python
#Release date comparison of all DC movies against Worldwide Gross
"""
This trend shows us people's sentiment towards the DC movies according to
their release year
According to the graph, a mixed general sentiment can be seen, since there is abrupt 
rise and fall in the graph
"""
plt.figure()
ax=sns.scatterplot(x="Release",y="Gross Worldwide", data=dc, hue="Gross Worldwide", size="Gross Worldwide", palette="magma").set(title="Release Date Comparison of All DC Movies")
```
